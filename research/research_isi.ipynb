# Comparison

```python
import numpy as np
from pathlib import Path
from src.utils.common import load_json
from src.utils.plotters import multiplot
from src.models.isi import ISI
```

```python
transmission_rate = {}
for location in optimizator.config.data.locations:
    d = load_json(Path(os.path.join(optimizator.config.path_to_results, f'ISI/{location}.json')))
    model = ISI(**d['val_based']['params'], N=df[df['location']==location]['population'].iloc[0])
    transmission_rate[location] = model.transmission_rate_approximation(np.arange(len(df[df['location']==location])))
```

```python
multiplot([df[df['location']==loc]['date'] for loc in optimizator.config.data.locations], 
          transmission_rate.values(),
          labels=optimizator.config.data.locations,
          ylabel='transmission rate',
          yscale='log',
          params_dict=manager.config.plots_default_params)
```
